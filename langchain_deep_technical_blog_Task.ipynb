# LangChain Deep Dive: Building Modular LLM Applications

This notebook explains LangChain concepts with examples and simplified implementations.

## Introduction

LangChain is a framework used to build applications powered by Large Language Models (LLMs).

It helps in:
- Connecting LLMs with external tools
- Managing workflows using chains
- Adding memory to applications

Without LangChain, building complex LLM applications becomes difficult.

# LLM Concept

```python
# Simulated LLM (no API needed)

def fake_llm(prompt):
    return f"LLM Response: {prompt}"

print(fake_llm("Explain LangChain"))
```

This simulates how an LLM responds to a prompt.

#  Prompt Template

```python
def prompt_template(topic):
    return f"Explain {topic} in simple terms"

print(prompt_template("LangChain"))
```

Prompt templates allow dynamic input into prompts.

#  Chains

```python
def chain(topic):
    prompt = prompt_template(topic)
    return fake_llm(prompt)

print(chain("LangChain"))
```

Chains connect multiple steps: Prompt → LLM → Output.

 # Memory

```python
memory = []

def chat(input_text):
    memory.append(input_text)
    return f"Memory: {memory}"

print(chat("Hi"))
print(chat("What is LangChain?"))
```

Memory stores previous interactions.

 # Agents

```python
def calculator_tool(expression):
    return eval(expression)

def agent(query):
    if "calculate" in query:
        return calculator_tool("25*4")
    return "I don't understand"

print(agent("calculate"))
```

Agents decide which tool to use.

#  Vector Store

```python
documents = ["LangChain is powerful", "It helps build AI apps"]

def search(query):
    return [doc for doc in documents if "LangChain" in doc]

print(search("LangChain"))
```

Vector stores help in retrieving relevant data.

## Architecture

User Input → Prompt → LLM → Chain → Agent → Output
